```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

X = train.drop(['Id','SalePrice'],axis=1)
Y = train['SalePrice']

test = test.drop(['Id'],axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state =42)
X_train, X_test, y_train, y_test
```

```python
#One hot encoding

X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# Ensuring both datasets have same dummies coloumns
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded,join='outer',axis=1,fill_value=0)
print("Training Data:")
print(X_train_encoded.head())
print("\nTesting Data:")
print(X_test_encoded.head())
```

```python
#Linear Regression 

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train_encoded,y_train)

y_pred = lr.predict(X_test_encoded)

mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print('mse;',mse)
print('r2;',r2)
```

```python
# save the model
import joblib
joblib.dump(lr,'house_price_model.pkl')
```

```python
from flask import Flask, request, jsonify
import pandas as pd
import joblib

# Create Flask application
app = Flask(__name__)

# Load the model and the encoder (if necessary)
model = joblib.load('house_price_model.pkl')

# Define the prediction route
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get JSON data from request
        json_data = request.get_json()
        
        # Convert JSON data to DataFrame
        features_df = pd.DataFrame(json_data, index=[0])
        
        # Perform any necessary preprocessing like you did before model training
        features_df_encoded = pd.get_dummies(features_df)
        
        # Ensure that all columns used in training are present in the DataFrame
        # Load your training encoded DataFrame columns for alignment
        # For demonstration, assuming 'train_columns.pkl' is saved column names from training
        train_columns = joblib.load('train_columns.pkl')  # Load the column names used during training
        features_df_encoded = features_df_encoded.reindex(columns=train_columns, fill_value=0)
        
        # Predict using the model
        prediction = model.predict(features_df_encoded)
        
        # Return the prediction
        return jsonify({'prediction': prediction.tolist()})
    
    except Exception as e:
        return jsonify({'error': str(e)}), 500

# Run the Flask application
if __name__ == '__main__':
    app.run(debug=True)
```
